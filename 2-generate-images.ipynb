```python
from PIL import Image, ImageOps,ImageEnhance
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = "C://Users//Abdullah//Desktop//idk.jpg"
image = Image.open(image_path).convert('L') 
```

```python
def display_image(image, title):
    plt.title(title)
    plt.axis('off')
    plt.imshow(image,cmap='gray')

display_image(image, "Original Image")
```

```python
def data_augmentation(image):
    augmented_images = []

    #1 Flip horizontally
    augmented_images.append(ImageOps.mirror(image))

    #2 Flip vertically
    augmented_images.append(ImageOps.flip(image))
    
    #3 Rotate by 90 degrees
    augmented_images.append(image.rotate(90))
    #4
    augmented_images.append(image.rotate(180))
    #5
    augmented_images.append(image.rotate(270))
    #6
    augmented_images.append(image.rotate(45))
    
    #7 Generate noisy image by adding Gaussian noise to the original image
    noisy_image = np.clip(np.array(image) + np.random.normal(0, 0.1, image.size[::-1]), 0, 255).astype(np.uint8)
    augmented_images.append(noisy_image)

    #8 Brighten the image
    augmented_images.append(ImageEnhance.Brightness(image).enhance(1.4))

    #9 Translation
    augmented_images.append(image.transform(image.size, Image.AFFINE, (1, 0, -50, 0, 1, -50)))
    #10 Translation
    augmented_images.append(image.transform(image.size, Image.AFFINE, (1, 0, -50, 0, 1, -50)))


    return augmented_images

# Apply data augmentation
augmented_images = data_augmentation(image)
```

```python
for i, augmented_image in enumerate(augmented_images):
    display_image(augmented_image, f"Augmented Image {i+1}")
    plt.pause(0.1)
```
